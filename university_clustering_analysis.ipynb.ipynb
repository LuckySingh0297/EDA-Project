# Import important liberaries

```python
import pandas as pd

import matplotlib.pyplot as plt
import seaborn as sns
```

# Upload Dataset

```python
UNIV_df = pd.read_excel(r"University_Clustering.xlsx")
UNIV_df.columns
```

```python
#Create a boxplot
sns.boxplot(x=UNIV_df["GradRate"])
plt.xlabel('GradRate Boxplot')
plt.title('GradRate')
plt.show()
```

# insight
Most universities ka graduation rate 70–90% range me hai

Few universities low performance dikha rahe hain (outliers)

Indicates ki majority institutions academically consistent hain

# “Business Meaning:”
Example:

This suggests that students targeting high graduation success should prefer more selective universities.

```python
# Create a histogram
plt.hist(UNIV_df['GradRate'], bins=5, color='green')
plt.xlabel('GradRate')
plt.ylabel('Frequency')
plt.title('Histogram of GradRate')
plt.show()
```

# Insight:

Data high values (80–100) ki taraf skewed hai

Most universities high graduation outcomes achieve kar rahe hain

Low graduation rare hai → quality generally high hai

```python
# Create a density plot

sns.kdeplot(UNIV_df['GradRate'], fill=True, color='green')
plt.xlabel('GradRate')
plt.ylabel('Density')
plt.title('Density Plot of GradRate')
plt.show()
```

```python
# Create scatter plot
sns.scatterplot(x='Accept', y='GradRate', data=UNIV_df)
plt.xlabel('Accept')
plt.ylabel('GradRate')
plt.title('Acceptance vs Gradrate scatter plot')
plt.grid()
plt.show()
```

```python
# Create Heatmap 
# define the numerical correlation matrix
numerical_df = UNIV_df.select_dtypes(include=['number'])

numerical_corr = numerical_df.corr()
# plot the correlation matrix using seaborn
sns.heatmap(numerical_corr, cmap='coolwarm', annot=True)
plt.show()
```

# Explanation:

Shows correlation between numerical variables

+1 = strong positive
-1 = strong negative
0 = no relation
# Insight:
Graduation rate strongly related hai kuch features se (jaise tuition / quality indicators)

Acceptance rate negatively correlated hai performance se
